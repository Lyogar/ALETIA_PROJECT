Now, I proceed to store all the records into a single annual table that combines all the movie records and earnings from each of the sources. For this purpose, I have created the following function.

```python
def create_table(dfn,dfb,dfk,dff):
    for n in range (200):
        dff.at[n,'Gross'] =  str(dfn['Domestic'][n])
        dff.at[n,'Movie'] =  str(dfn['Movie'][n])
    for n in range (200):
        dff.at[200+n,'Gross'] =  str(dfb['Domestic'][n])
        dff.at[200+n,'Movie'] =  str(dfb['Release Group'][n])
    dff['Gross'] = dff['Gross'].str.slice(start=1)
    dff = dff[dff['Gross'] != '']
    dff['Gross'] = dff['Gross'].str.replace(',', '').astype(int)
    col_kagg = [1, 13]
    kn = dfk.iloc[:, col_kagg]
    kn.columns = ['Movie','Gross']
    dff = pd.concat([dff, kn], axis=0, ignore_index=True)
    dff = dff.groupby('Movie').mean().reset_index()
    return dff
```

Una vez que se tienen los registros juntos en una tabla se procede a hacer una limpieza para quedarnos solo con las peliculas de las cuales se tiene registro de su género y su distribuidora, para lo cual cree varios filtros para determinar si los valores comparados se tratan de la misma pelicula, si es asi se hace un promedio de las gancias y se guarda un solo registro.
La siguiente función tiene cuatro filtros con los que espero eliminar duplicados.
La primera parte fusiona los registros que tienen una similitud de más del 80 por ciento en sus nombres.
La segunda parte une los registros que tienen exactamente el mismo valor de ganancias, ya que se trataria de la misma película.
La tercera parte une los registros que comparten similitudes entre sus títulos y sus ganancias ya que es altamente probable que se trate de la misma pelicula.
Por último se hace un filtro en el que se unen los títulos que son los mismos pero no cuentan con la palabra "The" al inicio.

Once the records are consolidated in a table, the next step is to perform a cleanup to retain only the movies for which we have records of both their genre and distributor. For this purpose, I have created several filters to determine if the compared values belong to the same movie. If so, the earnings are averaged, and only one record is saved. The following function includes four filters aimed at eliminating duplicates. The first part merges records that share more than an 80 percent similarity in their names. The second part merges records with exactly the same earnings, as they represent the same movie. The third part combines records that share similarities between their titles and earnings, as it is highly likely they represent the same movie. Finally, a filter is applied that merges titles that are the same but do not begin with the word "The."

```python
def filtr(df):
    from fuzzywuzzy import fuzz
    for i in range(1, len(df)):
        str1 = str(df.loc[i-1, 'Movie'])
        str2 = str(df.loc[i, 'Movie'])
        similitud = fuzz.ratio(str1, str2)
        if similitud >= 80:
            df.loc[i, 'Movie'] = str1       
    df = df.groupby('Movie').mean().reset_index()
    for i in range(1, len(df)):
        str1 = str(df.loc[i-1, 'Movie'])
        str2 = str(df.loc[i, 'Movie'])
        num1 = str(df.loc[i-1, 'Gross'])
        num2 = str(df.loc[i, 'Gross'])
        if num1 == num2:
            df.loc[i, 'Movie'] = str1
    df = df.groupby('Movie').mean().reset_index()
    for i in range(1, len(df)):
        str1 = str(df.loc[i-1, 'Movie'])
        str2 = str(df.loc[i, 'Movie'])
        num1 = str(df.loc[i-1, 'Gross'])[:3]
        num2 = str(df.loc[i, 'Gross'])[:3]                
        similitud = fuzz.ratio(str1, str2)
        if similitud >= 60 and num1 == num2:
            df.loc[i, 'Movie'] = str1
    df = df.groupby('Movie').mean().reset_index()
    for i in range(1, len(df)):
        str1 = str(df.loc[i, 'Movie'])[:3]
        if str1 == 'The':
            str1 = (df.loc[i, 'Movie'])[4:]
            for j in range(1, len(df)):
                str2 = str(df.loc[j, 'Movie'])
                if str1 == str2:
                    df.loc[i, 'Movie'] = str1
    df = df.groupby('Movie').mean().reset_index()
    return df
```

```python
def date_adj(df):
    from datetime import datetime
    original = "%b %d, %Y"
    modify = "%Y-%m-%d"
    for i in range(len(df)):
        release = df.loc[i, 'Release']
        fdatetime = datetime.strptime(release, original)
        df.loc[i, 'Release'] = fdatetime.strftime(modify)
    return df

index2000 = date_adj(index2000)
index2001 = date_adj(index2001)
index2002 = date_adj(index2002)
index2003 = date_adj(index2003)
index2004 = date_adj(index2004)
index2005 = date_adj(index2005)
index2006 = date_adj(index2006)
index2007 = date_adj(index2007)
index2008 = date_adj(index2008)
index2009 = date_adj(index2009)
index2010 = date_adj(index2010)
index2011 = date_adj(index2011)
index2012 = date_adj(index2012)
index2013 = date_adj(index2013)
index2014 = date_adj(index2014)
index2015 = date_adj(index2015)
index2016 = date_adj(index2016)
index2017 = date_adj(index2017)
index2018 = date_adj(index2018)
index2019 = date_adj(index2019)
index2020 = date_adj(index2020)
index2021 = date_adj(index2021)
index2022 = date_adj(index2022)
```
